```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("total_global_emissions.csv")
```

```python
print(df.head())
print(df.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="year", y="co2", marker="o")

plt.title("Global CO₂ Emissions Over Time", fontsize=16)
plt.xlabel("Year")
plt.ylabel("CO₂ Emissions (million tonnes)")
plt.tight_layout()

plt.savefig("global_co2_emissions.png", dpi=300)
plt.show()
```

```python
df = pd.read_csv("country_emissions_over_time.csv")
```

```python
print(df.head())
print(df.info())
```

```python
countries_to_plot = ["United States", "China", "India", "Germany", "Brazil"]
df_subset = df[df["country"].isin(countries_to_plot)]
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_subset, x="year", y="co2_running_total", hue="country", marker="o")

plt.title("Cumulative CO₂ Emissions by Country Over Time", fontsize=16)
plt.xlabel("Year")
plt.ylabel("CO₂ Emissions (Million Tonnes)")
plt.grid(True)
plt.tight_layout()

plt.savefig("cumulative_co2_by_country.png", dpi=300)
plt.show()
```

```python
df_increases = pd.read_csv("top_10_co2_increases.csv")

print(df_increases.head())
print(df_increases.info())
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df_increases, x="absolute_increase", y="country", palette="rocket")

plt.title("Top 10 Countries by Absolute Increase in CO₂ Emissions (Last 25 Years)", fontsize=16)
plt.xlabel("Absolute Increase in CO₂ Emissions (Million Tonnes)")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
df_decreases = pd.read_csv("top_10_co2_decreases.csv")

print(df_increases.head())
print(df_increases.info())
```

```python
df_decreases['absolute_decrease'] = df_decreases['absolute_increase'].abs()

plt.figure(figsize=(12, 6))
sns.barplot(data=df_decreases, x='absolute_decrease', y='country', palette='crest')

plt.title('Top 10 Countries by Absolute Decrease in CO₂ Emissions (Last 25 Years)', fontsize=16)
plt.xlabel('Absolute Decrease in CO₂ Emissions (Million Tonnes)')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
df_gdp = pd.read_csv("gdp_vs_emissions.csv")

print(df_gdp.head())
print(df_gdp.info())
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_gdp, x='gdp', y='co2', alpha=0.6)

plt.title('GDP vs CO₂ Emissions (Annual)', fontsize=16)
plt.xlabel('GDP (USD)')
plt.ylabel('Annual CO₂ Emissions (Million Tonnes)')
plt.tight_layout()
plt.show()
```

```python
corr = df_gdp[['gdp', 'co2']].corr()
print("Correlation between GDP and CO₂ emissions:")
print(corr)
```
