```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("prem_correlation_sql.csv")
```

```python
cols = ['total_spent', 'total_points', 'goal_difference', 'league_position']
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
df = df.dropna(subset=cols)
```

```python
def scatter_with_corr(x, y, data):
    corr = data[[x, y]].corr().iloc[0, 1]
    plt.figure(figsize=(8, 5))
    sns.regplot(data=data, x=x, y=y, scatter_kws={'s': 50, 'alpha': 0.7})
    plt.title(f"{x.replace('_', ' ').title()} vs {y.replace('_', ' ').title()}\n"
              f"Pearson r = {corr:.3f}", fontsize=13)
    plt.xlabel(x.replace('_', ' ').title())
    plt.ylabel(y.replace('_', ' ').title())
    plt.tight_layout()
    plt.show()
```

```python
scatter_with_corr('total_spent', 'total_points', df)
scatter_with_corr('total_spent', 'goal_difference', df)
scatter_with_corr('total_spent', 'league_position', df)
```
