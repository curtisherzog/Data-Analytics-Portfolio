```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
gdp = pd.read_csv('rank_gdp.csv')
social_support = pd.read_csv('rank_social_support.csv')
life_expectancy = pd.read_csv('rank_life.csv')
freedom = pd.read_csv('rank_freedom.csv')
generosity = pd.read_csv('rank_generocity.csv')
corruption = pd.read_csv('rank_corruption.csv')
```

```python
def plot_and_correlate(df, x_col, y_col, title):
    print(f"\n=== {title} ===")
    
    sns.scatterplot(data=df, x=x_col, y=y_col)
    plt.title(title)
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.show()
    
    corr_df = df[[x_col, y_col]].corr()
    print(corr_df)
    print(df.info())
    
    return corr_df.loc[x_col, y_col]
```

```python
correlations = {}
```

```python
correlations['GDP vs Happiness Ranking'] = plot_and_correlate(gdp, 'gdp_per_capita', 'rank', 'GDP vs Happiness Rank')
```

```python
correlations['Social Support vs Happiness Ranking'] = plot_and_correlate(social_support, 'social_support', 'rank', 'Social Support vs Happiness Rank')
```

```python
correlations['Healthy Life Expectancy vs Happiness Ranking'] = plot_and_correlate(life_expectancy, 'healthy_life_expectancy', 'rank', 'Healthy Life Expectancy vs Happiness Rank')
```

```python
correlations['Freedom vs Happiness Ranking'] = plot_and_correlate(freedom, 'freedom', 'rank', 'Freedom vs Happiness Rank')
```

```python
correlations['Generosity vs Happiness Ranking'] = plot_and_correlate(generosity, 'generosity', 'rank', 'Generosity vs Happiness Rank')
```
